```python
import mteb
from sentence_transformers import SentenceTransformer
```

```python
# Define both models
danish_model = SentenceTransformer("KennethTM/MiniLM-L6-danish-encoder")
english_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

# WikiClusteringP2P.v2

```python
# Get Danish STS tasks
tasks_wiki = mteb.get_tasks(tasks=["WikiClusteringP2P.v2"], languages=["dan"])
evaluation_wiki = mteb.MTEB(tasks=tasks_wiki)

results_wiki_danish = evaluation_wiki.run(danish_model, output_folder="results/danish")
results_wiki_english = evaluation_wiki.run(english_model, output_folder="results/english")
```

```python
for r in results_wiki_danish:
    print(f"Danish encoder - {r.task_name}: {r.scores}")

for r in results_wiki_english:
    print(f"English encoder - {r.task_name}: {r.scores}")
```

# SIB200ClusteringS2S

```python
tasks_sib = mteb.get_tasks(tasks=["SIB200ClusteringS2S"], languages=["dan"])
evaluation_sib = mteb.MTEB(tasks=tasks_sib)

results_danish_sib = evaluation_sib.run(danish_model, output_folder="results/danish")
results_english_sib = evaluation_sib.run(english_model, output_folder="results/english")
```

```python
for r in results_danish_sib:
    print(f"Danish encoder - {r.task_name}: {r.scores}")

for r in results_english_sib:
    print(f"English encoder - {r.task_name}: {r.scores}")
```
